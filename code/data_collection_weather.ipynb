```python
# importing dependendencies 

import pandas as pd
import requests
import numpy as np
import datetime

# keys
from config_weather import api_key, email
from config_weather1 import api_key_1, email
from config_weather2 import api_key_2, email
from config_weather3 import api_key_3, email
```

```python
# function to retrieve daily weather data

def get_daily_temp(city, date, includelocation, tp, fmt, key):

    url = "https://api.worldweatheronline.com/premium/v1/past-weather.ashx"

    payload = {
        "q": "Austin, Tx",
        "date": date,
        "includelocation": "yes",
        "tp": "24",
        "format": fmt,
        "key": api_key
    }

    response = requests.get(url, params=payload)
  
    if response.status_code == 200:

        res = response.json()["data"]["weather"][0]

        df = pd.DataFrame({"date":res["date"],
                           "avgtempC":res["avgtempC"],
                           "avgtempF":res["avgtempF"],
                           "winddirDegree":res["hourly"][0]["winddirDegree"],
                           "windspeedKmph":res["hourly"][0]["windspeedKmph"],
                           "windspeedMiles":res["hourly"][0]["windspeedMiles"],
                           "cloudcover":res["hourly"][0]["cloudcover"],
                           "humidity":res["hourly"][0]["humidity"],
                           "precipInches":res["hourly"][0]["precipInches"],
                           "precipMM":res["hourly"][0]["precipMM"],
                           "pressure":res["hourly"][0]["pressure"],
                           "precipMM":res["hourly"][0]["precipMM"],
                           "visibility":res["hourly"][0]["visibility"],
                           "precipMM":res["hourly"][0]["precipMM"],
                           "maxtempC":res["maxtempC"],
                           "maxtempF":res["maxtempF"],
                           "mintempC":res["mintempC"],
                           "mintempF":res["mintempF"],
                           "sunHour":res["sunHour"],
                           "uvIndex":res["uvIndex"],                       
                          }, 
                          index=[0])
        return df

    else:
        return None
```

```python
# extracting 2015 weather data

temp_2015_df = pd.DataFrame()

begin_year = datetime.date(2015, 1, 1)
end_year = datetime.date(2015, 12, 31)
one_day = datetime.timedelta(days=1)
next_day = begin_year

for day in range(0, 366): 
    # for potential leap year
    if next_day > end_year:
        break
        
    temp_2015_df = temp_2015_df.append(get_daily_temp(city="Austin,Tx",
                                                      date=next_day,
                                                      includelocation="yes",
                                                      tp="24",
                                                      fmt="json",
                                                      key=api_key))
        
    next_day += one_day
    
```

```python
# exporting 2015 weather data into a csv

temp_2015_df.to_csv("../data/daily_weather_2015_austin.csv", index=False)
```
